```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# data collection in a pandas frame

heart_dataset = pd.read_csv('/content/heart_disease_data.csv')
```

```python
heart_dataset.head()
```

```python
heart_dataset.shape
```

```python
# checking for any missing values in the dataset
heart_dataset.isnull().sum()
```

```python
heart_dataset.info()
```

```python
# So there are no missing values in the dataset , we can proceed :) )
heart_dataset.describe()
# we run this function to get an idea about the statistical measures of the dataset
```

```python
# checking the distribution of the output label
heart_dataset['target'].value_counts()
```

```python
# here , 1 will be considered as healthy and
# 0 as having disease
```

```python
# Splitting the target label and other data
X = heart_dataset.drop(columns='target',axis=1)
Y = heart_dataset['target']

print(X)
print(Y)
```

```python
# splitting the data into train and test
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
X_train.isnull().sum()
```

```python
# making ana training the model
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
# checking accuracy on train and test data
train_data_prediction = model.predict(X_train)
train_accuracy_score = accuracy_score(train_data_prediction,Y_train)
print(train_accuracy_score)
```

```python
# checking the model accuracy on the test data
test_data_prediction = model.predict(X_test)
test_accuracy_score = accuracy_score(test_data_prediction,Y_test)
print(test_accuracy_score)
```

```python
# Building a predictive system /
input_data = (69,0,3,140,239,0,1,151,0,1.8,2,2,2)
input_data_as_numpy_array = np.asarray(input_data)
# we'll be reshaping the new array to tell the function that we are predicting for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if prediction[0]==1 :
  print("The heart is healthy.")
else :
  print("The heart isn't healthy.")

print('Actual answer is : ', 1)
print('Actual heart condition is : healthy')
```
